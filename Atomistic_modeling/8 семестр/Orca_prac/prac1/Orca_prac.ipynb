```python
import pandas as pd
# import scikit-learn
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/energy_values.txt', sep=" ", header=None)
x = pd.read_csv('/content/x_values.txt', sep=" ", header=None)
```

```python
plt.plot(x, df)
```

```python
from scipy.interpolate import CubicSpline

cs = CubicSpline(x.to_numpy()[:, 0], df.to_numpy()[:, 0])
```

```python
x_smooth = np.linspace(min(x.to_numpy()[:, 0]), max(x.to_numpy()[:, 0]), 500)
y_smooth = cs(x_smooth)
```

```python
plt.plot(x, df)
plt.plot(x_smooth, y_smooth)
print(f'E min {min(y_smooth)}')
print(f'R min {x_smooth[np.argmin(y_smooth)]}')
print(f'Diss energy {min(y_smooth) - df.to_numpy()[:, 0][-1]}')
```
